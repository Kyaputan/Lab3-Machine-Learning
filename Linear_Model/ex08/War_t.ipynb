```python
X = df_Ground_vehicles.drop(["rb_win_rate"], axis = 1 )
y = df_Ground_vehicles["rb_win_rate"]
```

```python
X = USSR_df.drop(["rb_win_rate","nation","cls"], axis = 1 )
y = USSR_df["rb_win_rate"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score , mean_squared_error
from sklearn.tree import plot_tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
model = RandomForestRegressor(n_estimators=5)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
print(f'Mean Squared Error: {mse:.2f}')
```

```python
plt.scatter(range(len(y_pred)), y_pred, color="red", s=2, label='Predicted')  # กราฟสำหรับ y_pred
plt.scatter(range(len(y_test)), y_test, color="blue", s=5, label='Actual')  # กราฟสำหรับ y_test

# กำหนดชื่อแกน
plt.xlabel('Index')
plt.ylabel('Frags per Battle / Frags per Death')

# เพิ่ม legend
plt.legend()

# แสดงกราฟ
plt.show()
```
