# ตรวจว่าเท่ากันไหม เผื่อๆไว้ก่อน

```python
import os
import pandas as pd
from tqdm import tqdm


folder_path = "train/train"


for filename in sorted(os.listdir(folder_path)):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)

        # ตรวจสอบจำนวนแถว
        total_rows = len(df)
        remainder = total_rows % 480
        
        num_files = total_rows // 480
        
        if remainder == 0:
            print(f"{filename}: ✅ หาร 480 ลงตัว ({total_rows} แถว) -> สามารถสร้างได้ {num_files} ไฟล์")
        else:
            print(f"{filename}: ❌ ไม่ลงตัว ({total_rows} แถว, เหลือ {remainder} แถว) -> สามารถสร้างได้ {num_files} ไฟล์")
```

# แยกๆๆๆ ทำตาม test

```python
folder_path = "train/train"
output_folder = "train/split"

# ตรวจสอบว่าโฟลเดอร์ output มีอยู่หรือไม่ ถ้าไม่มีให้สร้างขึ้นมา
os.makedirs(output_folder, exist_ok=True)

# วนลูปอ่านไฟล์ทั้งหมดในโฟลเดอร์
for filename in tqdm(sorted(os.listdir(folder_path)), desc="Processing CSV Files"):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)

        # แบ่งข้อมูลเป็นชุดละ 480 แถว
        for i, chunk in enumerate(range(0, len(df), 480)):
            df_chunk = df.iloc[chunk:chunk + 480]

            # บันทึกไฟล์ใหม่
            output_file = os.path.join(output_folder, f"{filename.replace('.csv', '')}_{i+1}.csv")
            df_chunk.to_csv(output_file, index=False)

print("✅ แยกข้อมูลเสร็จสิ้น!")
```

# แยกตาม lable

```python
output_folder = "SSC/train" # โฟลเดอร์ที่จะบันทึกไฟล์ใหม่
split_folder = "train/split" # โฟลเดอร์ที่มีไฟล์ที่แยกแล้ว แต่ยังไม่แยกตาม label


labels = ['W', 'N', 'R']
for label in labels:
    os.makedirs(os.path.join(output_folder, label), exist_ok=True)
```

```python
if not os.path.exists(split_folder):
    print(f"โฟลเดอร์ {split_folder} ไม่พบ!")
else:
    # ใช้ tqdm เพื่อแสดง Progress Bar ขณะอ่านไฟล์
    for filename in tqdm(sorted(os.listdir(split_folder)), desc="Processing CSV Files"):
        if filename.endswith(".csv"):
            file_path = os.path.join(split_folder, filename)
            df = pd.read_csv(file_path)

            label_column = df.columns[-1]
            unique_labels = df[label_column].unique()

            # แยกข้อมูลตาม label
            for unique_label in unique_labels:
                df_label = df[df[label_column] == unique_label]

                # บันทึกไฟล์ใหม่ในโฟลเดอร์ที่สัมพันธ์กับ label
                if not df_label.empty:
                    output_file = os.path.join(output_folder, unique_label, f"{filename.replace('.csv', '')}_{unique_label}.csv")
                    df_label.to_csv(output_file, index=False)

print("✅ แยกข้อมูลตาม label เสร็จสิ้น!")
```

# ลบข้อมูลที่ไม่ใช้

```python
output_folder = "SSC/train"
labels = ['N', 'R', 'W']


for label in labels:
    label_folder = os.path.join(output_folder, label)
    print(f"กำลังลบคอลัมน์ Sleep_Stage จากไฟล์ในโฟลเดอร์ {label_folder}")
    
    if not os.path.exists(label_folder):
        print(f"โฟลเดอร์ {label_folder} ไม่พบ!")
        continue

    # ใช้ tqdm เพื่อแสดง Progress Bar ขณะอ่านไฟล์
    for filename in tqdm(sorted(os.listdir(label_folder)), desc=f"Processing {label} Files"):
        if filename.endswith(".csv"):
            file_path = os.path.join(label_folder, filename)
            df = pd.read_csv(file_path)

            if 'Sleep_Stage' in df.columns:
                df = df.drop(columns=['Sleep_Stage'])  
                df.to_csv(file_path, index=False)
            else:
                print(f"❌ ไม่พบคอลัมน์ Sleep_Stage ในไฟล์ {filename}")

print("✅ ลบคอลัมน์ Sleep_Stage จากทุกไฟล์เสร็จสิ้น!")
```

```python
import os
import pandas as pd
from glob import glob
from tqdm import tqdm
input_dir = 'SSC'
train_dir = input_dir + '/train'
df_paths = glob(os.path.join(train_dir, '*','*.csv'))  

data_list = []  # เก็บข้อมูลจากทุกไฟล์

for file_path in tqdm(df_paths):
    folder_name = os.path.basename(os.path.dirname(file_path))
    
    
    df = pd.read_csv(file_path)  # โหลด CSV
    
    # คำนวณ mean, std, max, min, sum สำหรับแต่ละ feature
    mean_vals = df.mean()
    std_vals = df.std()
    max_vals = df.max()
    min_vals = df.min()
    sum_vals = df.sum()
    
    # สร้าง Series ที่ประกอบด้วยสถิติต่าง ๆ และตั้งชื่อคอลัมน์
    stats_df = pd.DataFrame({
        'mean': mean_vals,
        'std': std_vals,
        'max': max_vals,
        'min': min_vals,
        'sum': sum_vals
    })
    
    # เปลี่ยนให้เป็นแถวเดียว
    stats_df = stats_df.T  # Transpose
    
    # ตั้งชื่อคอลัมน์ใหม่
    new_columns = [f"{col}_{stat}" for stat in ['mean', 'std', 'max', 'min', 'sum'] for col in df.columns]
    stats_df = stats_df.values.reshape(1, -1)  # Flatten the data
    stats_df = pd.DataFrame(stats_df, columns=new_columns)

    stats_df['label'] = folder_name
    
    data_list.append(stats_df)

# รวมทุกไฟล์เป็น DataFrame เดียว
df_final = pd.concat(data_list, ignore_index=True)

# เก็บเป็น CSV
df_final.to_csv('summary_stats_test.csv', index=False)

print(df_final.head())  # ดูตัวอย่างข้อมูล
```
