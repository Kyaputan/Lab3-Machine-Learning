```python
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
print("Loading data file now, this could take a while depending on file size")

df_non = pd.read_csv('Data_non_yolo.csv')
df_aug = pd.read_csv('Yolo_aug.csv')
df_ii = pd.read_csv('Yolo_ii.csv')
df_cc = pd.read_csv('CNN.csv')
```

```python
df_non.set_index('Epoch', inplace=True)
df_aug.set_index('Epoch', inplace=True)
df_ii.set_index('Epoch', inplace=True)
df_cc.set_index('Epoch', inplace=True)
```

```python
numerical_columns = df_non.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']).columns.tolist()
print("\nNumerical Columns:", numerical_columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_histograms(df, numerical_columns):
    for column in numerical_columns:
        plt.figure(figsize=(4, 2))
        sns.histplot(df[column], bins=30, kde=True, color='#ff8c00')  
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()


numerical_columns = ['mAP.50', 'mAP50-95', 'Precision', 'Recall','F1-Score']
plot_histograms(df_non, numerical_columns)
```

```python
correlation_matrix = df_non[numerical_columns].corr()
plt.figure(figsize=(4, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Numerical Columns')
plt.show()
```

```python
numerical_columns = df_aug.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64']).columns.tolist()
print("\nNumerical Columns:", numerical_columns)
```
